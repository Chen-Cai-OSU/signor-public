```python
import graphonlib
import numpy as np
```

```python
# first, we create a simple blockmodel graphon
w = graphonlib.graphons.make_hierarchical_blockmodel_graphon(1, [.3, .8])
```

```python
graphonlib.plot_graphon(w)
```

```python
# next, we generate a discrete edge probability matrix, p
grid = np.linspace(0, 1, 100)
p = graphonlib.evaluate(w, grid)
```

```python
# then we sample a random adjacency matrix
a = graphonlib.sample_edges(p)
```

```python
graphonlib.plot_adjacency(a)
```

```python
# now we smooth
# parameters:
#    h : neighborhood size parameter. Example: 0.3 means to include 
#        closest 30th percentile of neighbors in neighborhood
p_zhang = graphonlib.smoothing.zhang.smoother(a, h=0.3)
```

```python
graphonlib.plot_adjacency(p_zhang)
```
